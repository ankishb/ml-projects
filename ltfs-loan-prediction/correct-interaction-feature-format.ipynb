```python
import numpy as np
import pandas as pd
import glob
import os, gc
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import glob
cat_inter_2way = glob.glob('store_data/new_feat/cat_inter_2way/*')
cat_inter_3way = glob.glob('store_data/new_feat/cat_inter_3way/*')
int_inter_2way = glob.glob('store_data/new_feat/int_inter_2way/*')
int_inter_3way = glob.glob('store_data/new_feat/int_inter_3way/*')

len(cat_inter_2way), len(cat_inter_3way), len(int_inter_2way), len(int_inter_3way)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
total_feat = 0
feature_df = pd.DataFrame()
idx = 0
for file_no,file in enumerate(cat_inter_2way):
    if feature_df.shape[1] > 70:
        print(feature_df.shape, "==>", end=" ")
        file_name = 'store_data/new_feat/cat_inter_2way_'+str(idx)+'.csv'
        feature_df.iloc[:,:70].to_csv(file_name, index=None)
        feature_df = feature_df.iloc[:,70:]
        print(feature_df.shape)
        idx += 1
        
    feat = pd.read_csv(file)
    
    ####### Label Encoding ########
    le = LabelEncoder()
    feat.fillna(0, inplace=True)
    for col in feat.columns:
        try:
            feat[col] = le.fit_transform(feat[col])
        except:
            print(file, "==", col)
    ####### Label Encoding ########
    
    feature_df = pd.concat([feature_df, feat], axis=1)
    print(file_no, "==", file)
    
print(feature_df.shape, "==>", end=" ")
file_name = 'store_data/new_feat/cat_inter_2way_'+str(idx)+'.csv'
feature_df.to_csv(file_name, index=None)
print(feature_df.shape)
print("===========finish===========")
```
