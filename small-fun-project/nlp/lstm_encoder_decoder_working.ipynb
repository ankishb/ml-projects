```python
import numpy as np
from keras.models import Sequential
from keras.layers import LSTM, Dense, TimeDistributed, RepeatVector
```

```python
def generate_sequence(length, max_num):
    return np.random.randint(0,max_num, size=length)

generate_sequence(6,20)
```

```python
def one_hot_encode(seq, max_num):
    return np.array([np.eye(max_num)[i] for i in seq])

one_hot_encode(generate_sequence(3,5),5)
```

```python
def one_hot_decode(one_hotted_seq, max_num):
    return [np.argmax(i) for i in one_hotted_seq]

one_hot_decode(list(one_hot_encode(generate_sequence(3,5),5)), 5)
```

```python
def get_pair(n_in, n_out, max_num):
    in_seq = generate_sequence(n_in, max_num)
    out_seq = in_seq[:n_out]#[0] + [0 for _ in range(2)]
    print out_seq
    out_seq = [out_seq[0],0,0]
    print out_seq
#     print in_seq, out_seq
    x = one_hot_encode(in_seq, max_num)
    y = one_hot_encode(out_seq, max_num)
    
    x = x.reshape(1, x.shape[0], x.shape[1])
    y = y.reshape(1, y.shape[0], y.shape[1])
#     print x,y
    return x, y

x1, y1 = get_pair(3,1, 5)
print x1 
print y1
print x1.shape, y1.shape
```

```python
n_features = 5
in_time_steps = 3
out_time_steps = 1
model = Sequential()
model.add(LSTM(100, input_shape = (in_time_steps, n_features)))
model.add(RepeatVector(in_time_steps))
# return_sequences: Boolean. Whether to return the last output 
# in the output sequence, or the full sequence.
model.add(LSTM(100, return_sequences=True))
model.add(TimeDistributed(Dense(n_features, activation='softmax')))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['acc'])
model.summary()
```

```python
for epoch in range(1000):
    x,y = get_pair(in_time_steps, 3, n_features)
#     Verbose mode. 0 = silent, 1 = progress bar, 2 = one line per epoch.
    model.fit(x,y, epochs=1, verbose=2)
```

```python
total, correct = 100,0
for _ in range(total):
    x,y = get_pair(in_time_steps, 3, n_features)
    y_hat = model.predict(x, verbose=0)
    if np.array_equal(one_hot_decode(y[0], n_features), one_hot_decode(y_hat[0], n_features)):
        correct += 1
print 'Accuracy: %.2f%%' %(float(correct)/float(total)*100.0)
```

```python
# spot check some examples
for _ in range(10):
    x,y = get_pair(in_time_steps, out_time_steps, n_features)
    yhat = model.predict(x, verbose=0)
#     print y_hat
    print('Expected:', one_hot_decode(y[0],n_features), 'Predicted', one_hot_decode(yhat[0],n_features))
```
