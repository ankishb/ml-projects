```python
import pandas as pd
```

```python
# list('abcd') --> 'a','b','c','d'
df1 = pd.DataFrame({'a':range(6),
                    'b':[5,3,6,9,2,4]},
                    index = list('abcdef'))
df1
```

```python
df2 = pd.DataFrame({'c':range(4),
                    'd':[10,20,30, 40]}, 
                    index=list('abhi'))
df2
```

```python
# default inner join
# both right and left index should pass
df3 = pd.merge(df1, df2, left_index=True, right_index=True)
df3
```

```python
# default left join
df4 = df1.join(df2)
df4
```

```python
# default outer join
df5 = pd.concat([df1, df2], axis=1)
df5
```

```python
# join by _common_ columns: 'a', 'c'
pd.merge(df1, df2, on=['b'])

# don;t understand the problem
```
